In this Approach we are going to create an extra coloumn which contaion 1 if the corresponding value is missing or NA else 0



For example we have Age feature which has some missing values in it so we creat another column name it AGE_NA and fill in the loc 1 where value is NA in Age coloumn and fill remaing Row with 0 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
from sklearn.svm import SVC
```

```python
data = pd.read_csv('/Users/nandrajog/Downloads/titanic.csv',low_memory=False)
```

```python
croped_data = data[['Age','Fare','Survived']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(croped_data,
                                                 croped_data.Survived,train_size = 0.7,random_state=0)

print(X_train.shape)
```

```python
X_train['Age_NA'] = np.where(X_train['Age'].isnull(),1,0)
X_test['Age_NA'] = np.where(X_test['Age'].isnull(),1,0)
```

```python
X_train
```

```python
X_test
```

```python
X_train['Age'].fillna(X_train.Age.median(),inplace=True)
X_test['Age'].fillna(X_test.Age.median(),inplace=True)
```

Logistic regression is sensitive to the scale of the variable so it's better to use Std scaler which transform the variable with mean 0 varaince 1

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score
```

```python
std = StandardScaler()
```

```python
X_train_trimmed = X_train[['Age','Fare']]
X_test_trimmed = X_test[['Age','Fare',]]
```

```python
X_train_rescaled = std.fit_transform(X_train_trimmed)
X_test_rescaled = std.fit_transform(X_test_trimmed)
```

```python
clf_logistic_regression = LogisticRegression(C= 100,random_state=44)
print("Train with Median Imputation")
clf_logistic_regression.fit(X_train_rescaled,Y_train)
score = clf_logistic_regression.score(X_train_rescaled,Y_train)
print("Score with Median Imputation {}".format(score))
predictions_train= clf_logistic_regression.predict_proba(X_train_rescaled)[:,1]
print("Roc_auc_score with Median Imputation  on Train set {}".format(roc_auc_score(Y_train,predictions_train)))

predictions= clf_logistic_regression.predict_proba(X_test_rescaled)[:,1]
print("Roc_auc_score with Median Imputation  on Test set {}".format(roc_auc_score(Y_test,predictions)))
```

```python
X_train_trimmed = X_train[['Age','Fare','Age_NA']]
X_test_trimmed = X_test[['Age','Fare','Age_NA']]
```
